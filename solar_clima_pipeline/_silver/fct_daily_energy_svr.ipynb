```python
spark.sql("CREATE SCHEMA IF NOT EXISTS  silver")
```

```python
df = spark.table("workspace.bronze.solar_production")
display(df)
```

```python
from pyspark.sql.functions import posexplode, to_date, regexp_extract, col

#  CRIA TABELA SILVER DADOS DIARIOS DE ENERGIA SOLAR
df_daily = df.select(
    to_date(regexp_extract("file_name", '(\\d{4}-\\d{2}-\\d{2})', 1),'yyyy-MM-dd').alias("date"),
    col("duration").cast("long"),
    col("total").cast("double"),
    col("co2").cast("double"),
    col("max").cast("double")
)
display(df_daily)

df_daily.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver.fct_daily_kwh")
```

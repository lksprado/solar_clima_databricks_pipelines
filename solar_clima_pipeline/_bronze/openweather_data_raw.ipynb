```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, FloatType, LongType

schema = StructType([
    StructField("lat", DoubleType(), True),
    StructField("lon", DoubleType(), True),
    StructField("tz", StringType(), True),
    StructField("date", StringType(), True),
    StructField("units", StringType(), True),
    StructField("cloud_cover", StructType([
        StructField("afternoon", DoubleType(), True)
    ]), True),
    StructField("humidity", StructType([
        StructField("afternoon", DoubleType(), True)
    ]), True),
    StructField("precipitation", StructType([
        StructField("total", DoubleType(), True)
    ]), True),
    StructField("temperature", StructType([
        StructField("min", DoubleType(), True),
        StructField("max", DoubleType(), True),
        StructField("afternoon", DoubleType(), True),
        StructField("night", DoubleType(), True),
        StructField("evening", DoubleType(), True),
        StructField("morning", DoubleType(), True)
    ]), True),
    StructField("pressure", StructType([
        StructField("afternoon", DoubleType(), True)
    ]), True),
    StructField("wind", StructType([
        StructField("max", StructType([
            StructField("speed", DoubleType(), True),
            StructField("direction", DoubleType(), True)
        ]), True)
    ]), True)
])

df = spark.read.format(
    "json"
).schema(
    schema
).load(
    "s3://openweatherbrz/bronze/"
)
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.weather_results")
display(df)
```
